### How many violations are considered critical ?

```python
critical = y_train.value_counts()
critical = critical[:10,]

plt.figure(figsize=(4,6))
sns.barplot(critical.index, critical.values, alpha=0.8)
plt.ylabel('Count of Violations', fontsize=20)
plt.xlabel('Violation Is Considered Critical', fontsize=16) 
plt.show()
```

* About half of all recorded violations are Critical Flags. This is dangerous since it is an indicator for how likely it is for a customer to get ill after eating at establishments who have critical violations.
